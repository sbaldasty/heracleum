```python
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.ticker import MaxNLocator
from src._poisoneffect import N_HONEST_CLIENTS
```

```python
pe_df = pd.read_csv('../out/poisoneffect.csv')
nb_df = pd.read_csv('../out/normball.csv')
```

```python
subdf = pe_df[pe_df['attack'] == 'Scaling (factor=2.5)'][pe_df['defense'] == 'No defense']
plt.plot(subdf['n_corrupt_clients'], subdf['accuracy'], label='Scaling (factor=2.5) / No defense')

subdf = pe_df[pe_df['attack'] == 'Scaling (factor=2.5)'][pe_df['defense'] == 'Norm ball']
plt.plot(subdf['n_corrupt_clients'], subdf['accuracy'], label='Scaling (factor=2.5) / Norm ball defense')

plt.plot(nb_df['n_corrupt_clients'], nb_df['accuracy'], label='Calibrated scaling attack / Norm ball defense')

plt.title(f'CIFAR-10 data set with {N_HONEST_CLIENTS} honest clients (no defense)')
plt.xlabel('# additional corrupt clients')
plt.ylabel('Model accuracy')
plt.legend()
plt.tight_layout()
plt.gca().xaxis.set_major_locator(MaxNLocator(integer=True))
plt.show()
```
